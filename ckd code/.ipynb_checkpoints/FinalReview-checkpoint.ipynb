```python
import warnings
warnings.filterwarnings("ignore")
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import seaborn as sns
```

```python
df = pd.read_csv('kidney_disease.csv')
df.head()
```

```python
# Impuing Null values

from sklearn.impute import SimpleImputer
imp_mode = SimpleImputer(missing_values=np.nan, strategy='most_frequent')

df_imputed=pd.DataFrame(imp_mode.fit_transform(df))
df_imputed.columns=df.columns
df_imputed
```

```python
df_imputed.isnull().sum()
```

```python
# Finding Unique values in the columns

for i in df_imputed.columns:
    print("*************************************************",i,"************************************************************")
    print()
    print(set(df_imputed[i].tolist()))
    print()
```

```python
print(df_imputed["rc"].mode())
print(df_imputed["wc"].mode())
print(df_imputed["pcv"].mode())
```

```python
df_imputed["classification"]
```

```python
df_imputed["classification"]=df_imputed["classification"].apply(lambda x:'ckd' if x=="ckd\t"  else x)

df_imputed["cad"]=df_imputed["cad"].apply(lambda x:'no' if x=="\tno"  else x)

df_imputed["dm"]=df_imputed["dm"].apply(lambda x:'no' if x=="\tno"  else x)
df_imputed["dm"]=df_imputed["dm"].apply(lambda x:'yes' if x=="\tyes"  else x)
df_imputed["dm"]=df_imputed["dm"].apply(lambda x:'yes' if x==' yes'  else x)

df_imputed["rc"]=df_imputed["rc"].apply(lambda x:'5.2' if x=='\t?'  else x)

df_imputed["wc"]=df_imputed["wc"].apply(lambda x:'9800' if x=='\t6200'  else x)
df_imputed["wc"]=df_imputed["wc"].apply(lambda x:'9800' if x=='\t8400'  else x)
df_imputed["wc"]=df_imputed["wc"].apply(lambda x:'9800' if x== '\t?'  else x)

df_imputed["pcv"]=df_imputed["pcv"].apply(lambda x:'41' if x=='\t43'  else x)
df_imputed["pcv"]=df_imputed["pcv"].apply(lambda x:'41' if x== '\t?'  else x)
```

```python
# Finding Unique values in the columns

for i in df_imputed.columns:
    print("*************************************************",i,"************************************************************")
    print()
    print(set(df_imputed[i].tolist()))
    print()
```

```python
# Label encoding to convert categorical values to numerical
from sklearn import preprocessing

df_enco=df_imputed.apply(preprocessing.LabelEncoder().fit_transform)
df_enco
```

```python
df_enco.to_csv("Kidney_Disease_Pre-Processed.csv")
```

```python
# Finding Correlations

import matplotlib.pyplot as plt

plt.figure(figsize=(20,20))
corr=df_enco.corr()
sns.heatmap(corr,annot=True)
```

```python
# Seperate independent and dependent variables and drop the ID column
x=df_enco.drop(["id","classification"],axis=1)
y=df_enco["classification"]
```

```python
scaler=MinMaxScaler((-1,1))
x=scaler.fit_transform(x)
```

```python
x
```

```python
from sklearn.decomposition import PCA

pca = PCA(.95)
X_PCA=pca.fit_transform(x)

print(x.shape)
print(X_PCA.shape)
```

```python

X_train, X_test, y_train, y_test = train_test_split(X_PCA,y, test_size=0.3,random_state=101)
```

```python
x_values = []
y_values = []
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt
```

## Naive Bayes 

```python
from sklearn.naive_bayes import GaussianNB
naive = GaussianNB()
naive.fit(X_train, y_train)
y_test_nb  =  naive.predict(X_test)
y_train_nb  =  naive.predict(X_train)
acc_train_nb = accuracy_score(y_train,y_train_nb)
acc_test_nb = accuracy_score(y_test,y_test_nb)
# print("Naive Bayes: Accuracy on test Data: {:.5f}".format(acc_test_nb))

# print the accuracy value
print("Naive Bayes: Accuracy on test Data: {:.2f}%".format(acc_test_nb * 100))
print("Naive Bayes: Accuracy on train Data: {:.2f}%".format(acc_train_nb * 100))
#format(acc_train_nb * 100, ".2f")
y_values.append(float(format(acc_test_nb * 100, ".2f")))
x_values.append("Naive Bayes")
```

##  Decision Tree

```python
# Decision Tree model 
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# instantiate the model 
tree = DecisionTreeClassifier(max_depth=3, min_samples_split=10)

# fit the model 
tree.fit(X_train, y_train)

# predicting the target value from the model for the samples
y_test_tree = tree.predict(X_test)
y_train_tree = tree.predict(X_train)

# calculate accuracy on train and test data
acc_train_tree = accuracy_score(y_train, y_train_tree)
acc_test_tree = accuracy_score(y_test, y_test_tree)

# print the accuracy value
print("Decision Tree Accuracy on test data: {:.2f}%".format(acc_test_tree * 100))
print("Decision Tree Accuracy on train data: {:.2f}%".format(acc_train_tree * 100))

y_values.append(float(format(acc_test_tree * 100, ".2f")))
x_values.append("Decision Tree")
```

## KNN

```python
#KNN Model


from sklearn.neighbors import KNeighborsClassifier

# create a KNN object with number of neighbors as 3
knn = KNeighborsClassifier(n_neighbors = 3)

# fit the model on training data
knn.fit(X_train, y_train)

# predict the target value for test and train data using the trained model
y_test_knn = knn.predict(X_test)
y_train_knn = knn.predict(X_train)

# calculate accuracy on train and test data
acc_train_knn = accuracy_score(y_train,y_train_knn) * 100
acc_test_knn = accuracy_score(y_test,y_test_knn) * 100

# print the accuracy value
print("KNN: Accuracy on test Data: {:.2f}%".format(acc_test_knn))
print("KNN: Accuracy on train Data: {:.2f}%".format(acc_train_knn))


y_values.append(float(format(acc_test_knn , ".2f")))
x_values.append("KNN")
```

## Random Forest 

```python

from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# set random seed for reproducibility
np.random.seed(42)

# instantiate the model 
rnd = RandomForestClassifier(n_estimators=5, min_samples_split=20, min_impurity_decrease=0.01, max_features=0.5)

# fit the model 
rnd.fit(X_train, y_train)

# predicting the target value from the model for the samples
y_test_rnd = rnd.predict(X_test)
y_train_rnd = rnd.predict(X_train)

# calculate accuracy on train and test data
acc_train_rnd = accuracy_score(y_train, y_train_rnd)
acc_test_rnd = accuracy_score(y_test, y_test_rnd)

# print the accuracy value
print("Accuracy on test data: {:.2f}%".format(acc_test_rnd * 100))
print("Accuracy on train data: {:.2f}%".format(acc_train_rnd * 100))

y_values.append(float(format(acc_test_rnd * 100, ".2f")))
x_values.append("Random Forest")
```

```python
from sklearn import svm
sv = svm.SVC(kernel='linear') 
sv.fit(X_train, y_train)

# predicting the target value from the model for the samples
y_test_sv = sv.predict(X_test)
y_train_sv = sv.predict(X_train)

# calculate accuracy on train and test data
acc_train_sv = accuracy_score(y_train, y_train_sv)
acc_test_sv = accuracy_score(y_test, y_test_sv)

# print the accuracy value
print("Accuracy on test data: {:.2f}%".format(acc_test_sv * 100))
print("Accuracy on train data: {:.2f}%".format(acc_train_sv * 100))

y_values.append(float(format(acc_test_sv * 100, ".2f")))
x_values.append("SVM")
```

```python
import matplotlib.pyplot as plt

# Data
#x_values = ['Naive Bayes', 'Decision Tree', 'KNN', 'Random Forest',"SVM"]
#y_values = [100.0, 99.17, 97.51, 99.17,99.17]

# Create a bar graph
plt.bar(x_values, y_values)

# Add title and axis labels
plt.title("Accuracy Scores")
plt.xlabel("Classifier")
plt.ylabel("Accuracy (%)")

# Set the y-axis limits and ticks
plt.ylim(97, 100)
plt.yticks([i/10 for i in range(970, 1001, 3)], [str(i/10) for i in range(970, 1001, 3)])

# Add labels for each bar
for i, v in enumerate(y_values):
    plt.text(i, v + 0.2, f"{v:.2f}%", ha='center')

# Show the graph
plt.show()
```
